```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
full_census_table=pd.read_csv("/home/ashly/Desktop/FDS/Lab Assignments/Lab FDS 7-10-20/CSV Files/census.csv")
partial_census_table=full_census_table[['SEX', 'AGE', 'POPESTIMATE2010', 'POPESTIMATE2014']]
us_pop=partial_census_table.rename(columns={'POPESTIMATE2010': '2010','POPESTIMATE2014':'2014'})
us_pop=us_pop.drop(columns="2010")
```

```python
#Finding Female Gender Ratio
females = us_pop.loc[np.logical_and(us_pop['SEX']==2,us_pop["AGE"]!=999)]
females
```

```python
males=us_pop.loc[np.logical_and(us_pop['SEX']==1,us_pop["AGE"]!=999)]
males
```

```python
m1=np.array(males['2014'])
m1
f1=np.array(females['2014'])
f1
ratios=f1/m1
ratios
ratios_table=pd.DataFrame(males["AGE"])
ratios_table['2014 F:M RATIO']=ratios
ratios_table
```

```python
males.loc[np.logical_and(males["AGE"]>=98,males["AGE"]<100)]
```

```python
females.loc[np.logical_and(females["AGE"]>=98,females["AGE"]<100)]
```

```python
plt.plot(ratios_table['AGE'],ratios_table['2014 F:M RATIO'])
plt.xlabel("2014 F:M RATIO")
plt.ylabel("AGE")
```
