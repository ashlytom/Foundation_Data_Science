```python
census=pd.read_csv('/home/ashly/Desktop/FDS/Lab Assignments/Lab FDS 7-10-20/CSV Files/census.csv')
census=census.drop(columns=['ESTIMATESBASE2010','CENSUS2010POP','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2015'])
census=census.rename(columns={'POPESTIMATE2010':'2010','POPESTIMATE2014':'2014'})
children=census.loc[census['SEX']==0]
children=children.loc[children.AGE<=18].drop(columns='SEX')
children
```
