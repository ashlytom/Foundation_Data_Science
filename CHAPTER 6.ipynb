### 6.3 POPULATION TRENDS

```python
full_census_table=pd.read_csv("/home/ashly/Desktop/FDS/Lab Assignments/Lab FDS 7-10-20/CSV Files/census.csv")
full_census_table.head()
```

```python
partial_census_table=full_census_table[['SEX', 'AGE', 'POPESTIMATE2010', 'POPESTIMATE2014']]
partial_census_table
```

```python
us_pop=partial_census_table.rename(columns={'POPESTIMATE2010': '2010','POPESTIMATE2014':'2014'})
us_pop
```

```python
cen=us_pop['2014'] - us_pop['2010']
cen
```

```python
change=(cen/us_pop['2010'])
percent_change=change.apply(abs)
percent_change=percent_change.apply(conv_to_perc)
percent_change
```

```python
us_pop['Change']=cen
us_pop['Percent Change']=percent_change
```

```python
p=us_pop.Change.apply(abs)
us_pop['Change']=p
us_pop
```

```python
us_pop.sort_values(by="Change",ascending=False)
```

#### 6.4 TRENDS IN GENDER

```python
us_pop1=us_pop.drop(columns=['Change','Percent Change'])
us_pop1
```

```python
e=us_pop1.loc[np.logical_and(us_pop1['SEX']==0,np.logical_and(us_pop1['AGE']>=97,us_pop1['AGE']<101))]
e
```

```python
us_pop1_2014=us_pop1.drop(columns="2010")
all_ages=us_pop1_2014.loc[us_pop1_2014.AGE==999]
all_ages
```

```python
t=all_ages['2014']/all_ages.iloc[0,2]
t=pd.DataFrame(t*100)
all_ages['Proportional']=t
all_ages

def to_conv(val):
    f=str(val)
    f=f+"%"
    return f

all_ages['Proportional']=all_ages['Proportional'].apply(to_conv)
all_ages
    
```
